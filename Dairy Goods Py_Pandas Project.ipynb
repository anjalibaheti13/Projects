1. Does number of cows in a farm has any dependency on area acquired?

```python
#let us first calculate their aggregate value:-

gb = df.groupby("Farm Size").agg({"Number of Cows":"sum", "Total Land Area (acres)":"sum"})
gb
```

```python
#Correlation to check dependency:-

c = gb.corr()
c
```

Yes, Number of cows in a farm has dependency on Total Land Area, as correlation is near to +1.

2. What is the average expiry days of each product?

```python
#to convert it into datetime --

df[["Production Date" ,	"Expiration Date"]] = df[["Production Date" , "Expiration Date"]].apply(pd.to_datetime)
df.info()
```

```python
#New column shelf life:-

df["Shelf Life"] = (abs(df["Production Date"] - df["Expiration Date"]))   #abs for absolute +ve value.
df.head()
```

```python
#aggregation:-

gb = df.groupby("Product Name").agg({"Shelf Life":"mean"})
gb
```

3. Mention the name of the products with max and min shelf life.

```python
#agrregation:-

gby1 = df.groupby("Product Name").agg({"Shelf Life": "min"})
gby1

gby2 = df.groupby("Product Name").agg({"Shelf Life": "max"})
gby2
```

```python
#merge:-

merge = pd.merge(left = gby1, right = gby2, on = "Product Name", how = "outer")
merge
```

4. Summarize the data on the basis of brands and land acquired by them,cows and the revenue generated by them. 

```python
gb = df.groupby("Brand").agg({"Total Land Area (acres)": "sum", "Number of Cows": "sum", "Approx. Total Revenue(INR)": "sum" })
gb
```

5. What is the average shelf life for different storage conditions? 

```python
gb = df.groupby("Storage Condition").agg({"Shelf Life": "mean"})
gb
```
